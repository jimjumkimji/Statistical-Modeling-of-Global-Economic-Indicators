## 5. Missing Values Analysis

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# แยก numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# เอา Year ออก (ไม่ต้อง aggregate)
numeric_cols.remove('Year')

numeric_cols
```

```python
df_agg = (
    df
    .groupby('Country Name')[numeric_cols]
    .median()
    .reset_index()
)

df_agg.head()
```

```python
df_agg = df_agg.merge(
    df_region,
    left_on='Country Name',
    right_on='CountryName',
    how='left'
)

df_agg.head()
```

```python
target_col = 'GNI per capita, PPP (current international $)'

df_agg = df_agg.dropna(subset=[target_col])

df_agg[target_col].isnull().sum()
```

```python
# หา numeric columns
numeric_cols = df_agg.select_dtypes(include=[np.number]).columns.tolist()

# เอา target ออก
numeric_cols.remove(target_col)

# หา column ที่ missing ทั้งหมด
all_missing_cols = [
    col for col in numeric_cols
    if df_agg[col].isna().all()
]

all_missing_cols
```

```python
df_agg = df_agg.drop(columns=all_missing_cols)
```

```python
from sklearn.impute import SimpleImputer

feature_cols = df_agg.select_dtypes(include=[np.number]).columns.tolist()
feature_cols.remove(target_col)

imputer = SimpleImputer(strategy='median')
df_agg[feature_cols] = imputer.fit_transform(df_agg[feature_cols])
```

## 2.3 Visualization Dashboard

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(
    df_agg['GNI per capita, PPP (current international $)'],
    bins=30
)
plt.xlabel('GNI per capita, PPP (current international $)')
plt.ylabel('Number of countries')
plt.title('Distribution of GNI per capita (PPP)')
plt.tight_layout()
plt.show()
```

```python
# Import NumPy for numerical operations
import numpy as np

df_agg['log_GNI'] = np.log(
    df_agg['GNI per capita, PPP (current international $)']
)

plt.figure(figsize=(6,4))
plt.scatter(
    df_agg['School enrollment, tertiary (% gross)'],
    df_agg['log_GNI'],
    alpha=0.7
)
plt.xlabel('School enrollment, tertiary (% gross)')
plt.ylabel('log(GNI per capita)')
plt.title('Relationship between Tertiary Education and GNI')
plt.tight_layout()
plt.show()
```

```python
# Count number of countries in each region
region_counts = df_agg['Region'].value_counts()

plt.figure(figsize=(7,4))
plt.bar(region_counts.index, region_counts.values)
plt.xlabel('Region')
plt.ylabel('Number of countries')
plt.title('Number of Countries by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Prepare time-series data for visualization
ts = (
    df.groupby('Year')[[
        'GNI per capita, PPP (current international $)',
        'School enrollment, tertiary (% gross)'
    ]]
    .median()
)

plt.figure(figsize=(7,4))
plt.plot(ts.index, ts['GNI per capita, PPP (current international $)'], label='Median GNI')
plt.plot(ts.index, ts['School enrollment, tertiary (% gross)'], label='Median Tertiary Enrollment')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Global Trends of GNI and Tertiary Education (2017–2021)')
plt.legend()
plt.tight_layout()
plt.show()
```
